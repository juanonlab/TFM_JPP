```python
# Juan Pardo Palazón

# Importacion de librerias
import numpy as np
import pandas as pd
import datetime
```

```python
# Constantes
ROOT_DIRECTORY = 'D:/anaconda/data/'
AFTER_DATA = '2014-02-01'
BEFORE_DATA = '2015-07-01'
FILE_BASE = 'CLEAN_House' 
WINDOW_SIZE = 30
FIELD = 'Aggregate'
DECIMALS = 2
```

```python
def getDfFiltered(file, column):
    """Filtrar por fechas dadas y seleccionar la columna a tratar (pasada por parámetro)
     
    Argumentos:
    file -- fichero de entrada 
    column -- columna a seleccionar
    """
    # Leer fichero origen, seleccionar columna consumo total ()
    house_data = pd.read_csv(ROOT_DIRECTORY + file+'.csv', delimiter = ',', parse_dates=[0], index_col=0)
    house_data = house_data[[column]]
    
    # Filtrardo de fechas. Rango de fechas con menor rango datos faltantes
    df_filtered = house_data[(house_data.index > AFTER_DATA) & (house_data.index < BEFORE_DATA)]
    return df_filtered
```

```python
def dfTransformFile(file, column):

    # Obtener data frame filtrado por fechas y con la columna pasada por parametro    
    df_filtered = getDfFiltered(file, column)
    
    # Se agrupa por rangos de 6 horas en un día
    dt_filtered_resample24H = df_filtered.resample('24H').mean()
   
    return dt_filtered_resample24H
```

```python
def fixCleanData(dt_filtered):

    # Se seleccionan aquellos valores del rango hora
    # pasado por parámetro (0, 6, 12 ó 18)
    # dt_filtered_Range = dt_filtered[dt_filtered_6H.Hour == rango]
    
    # Limpieza de NA
    dt_filtered = dt_filtered.interpolate(method ='linear', limit_direction ='forward')
    
    return dt_filtered
```

```python
def generateWeekFile(dt_filtered_24H, field):

    # Limpiar datos
    dataToFilter = fixCleanData(dt_filtered_24H)

    # Filtrar por hora y obtener por semana
    dt_filtered_24H_Week = dataToFilter.rolling(window=WINDOW_SIZE, center=True).mean().resample("W").mean().apply(lambda x: round(x, DECIMALS))

    return dt_filtered_24H_Week
```

```python
# Generar los ficheros
for x in range(1, 22):
    if x != 14:
        file = FILE_BASE + str(x)
        dt_filtered = dfTransformFile(file, FIELD)
        dt_filtered_Week  = generateWeekFile(dt_filtered, FIELD)
        # El hogar 6 no dispone datos del primer mes por eso se 
        # recoge desde la posición 7
        if x != 6:
            dt_filtered_Week.iloc[2:-2, :].to_csv(file+'_week'+'agg.csv', sep=',', encoding='utf-8')
        else:
            dt_filtered_Week.iloc[7:-2, :].to_csv(file+'_week'+'agg.csv', sep=',', encoding='utf-8')
        
```
